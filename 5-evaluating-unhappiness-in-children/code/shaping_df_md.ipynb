```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_feather('../data/h201.feather')
deps = pd.read_csv('../data/Health Status Variables.csv')
```

```python
df.loc[df['AGELAST'] <= 17]
```

```python
our_vars = [col for col in deps['VARIABLE'][30:97]]
our_vars.append('AGELAST')
```

```python
our_df = df.loc[(df['AGELAST'] <= 17) & (df['AGELAST'] >= 5), our_vars]
```

```python
our_df = our_df.loc[(our_df['UNHAP42'] >= 0) & (our_df['UNHAP42'] <= 4), :]
```

```python
our_df['UNHAP42'] = our_df['UNHAP42'].map({0:0, 1:1, 2:1, 3:1, 4:1})
```

```python
our_df['UNHAP42'].value_counts(normalize=True)
```

```python
our_df.to_csv('../data/semi_clean_df.csv', index=False)
```
